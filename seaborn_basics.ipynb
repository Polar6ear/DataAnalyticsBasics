```python
flight_df = sns.load_dataset('flights')
flight_df.head()
```

```python
flight_df = flight_df.pivot(index='month', columns='year', values='passengers')
flight_df.head()
```

```python
plt.figure(figsize=(12, 6))
ax=sns.heatmap(flight_df)
ax
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

grid_kws = {"height_ratios": (.85, .05), "hspace": .3}
f, (ax, cbar_ax) = plt.subplots(2, 1, gridspec_kw=grid_kws)

sns.heatmap(
    flight_df,
    ax=ax,
    cbar_ax=cbar_ax,
    cbar_kws={"orientation": "horizontal"}
)

plt.show()
```
